```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver = webdriver.Chrome()
```

```python
#Opening the naukri page on automated chrome browser
driver.get(' https://www.naukri.com/')
```

```python
# entring  job-title, job-location, company name, experience required as required in the question
designation = driver.find_element(By.CLASS_NAME, 'suggestor-input' ) # Makes the cursor comes inside the input field(box)
designation.send_keys('Data Scientist') # enter/input 'Data Analyst' into the box
```

```python
location = driver.find_element(By.XPATH, '/html/body/div/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Delhi/NCR')
```

```python
Dropdown = driver.find_element(By.CLASS_NAME, 'dropArrowDD')
Dropdown.click()

value = driver.find_element(By.XPATH, '//ul[@class="dropdown "]/li[5]/div/span')
value.click()
```

```python
search = driver.find_element(By.CLASS_NAME, 'qsbSubmit')
search.click()
```

```python
job_title = []
job_location = []
company_Name = []
experience_req = []
```

```python
#scraping job title from the given page
title_tag = driver.find_elements(By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tag[:10]:
    title = i.text
    job_title.append(title)
    
#scraping job location from the given page
location_tag = driver.find_elements(By.XPATH, '//span[@class="loc-wrap ver-line"]/span/span')
for i in location_tag[:10]:
    location = i.text
    job_location.append(location)
    
#scraping job company from the given page
company_tag = driver.find_elements(By.XPATH, '//div[@class=" row2"]/span/a[1]')
for i in company_tag[:10]:
    company = i.text
    company_Name.append(company)
    
#scraping job experience from the given page
experience_tag = driver.find_elements(By.XPATH, '//span[@class="exp-wrap"]/span/span')
for i in experience_tag[:10]:
    experience = i.text
    experience_req.append(experience)
```

```python
print(len(job_title), len(job_location), len(company_Name), len(experience_req))
```

```python
import pandas as pd

df = pd.DataFrame({'Title': job_title, 'Location':job_location, 'Company Name': company_Name, 'Experince': experience_req })

df
```

```python
driver = webdriver.Chrome()

driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART') #Opening the naukri page on automated chrome browser
```

```python
rating_given = []
review_summ = []
full_review = []


start = 0
end = 10
for page in range(start, end):
    rating_tag = driver.find_elements(By.XPATH, '//DIV[@class="_3LWZlK _1BLPMq"]')
    for i in rating_tag:
        rating_given.append(i.text)
    next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]/span') #botton to move to next page
    next_button.click()
    time.sleep(3)
    
for page in range(start, end):
    review_tag = driver.find_elements(By.CLASS_NAME, '_2-N8zT')
    for i in review_tag:
        review_summ.append(i.text)
    next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]/span') #botton to move to next page
    next_button.click()
    time.sleep(3)
    
for page in range(start, end):
    fullreview_tag = driver.find_elements(By.XPATH, '//div[2][@class="row"]/div/div/div')
    for i in fullreview_tag:
        full_review.append(i.text)
    next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]/span') #botton to move to next page
    next_button.click()
    time.sleep(3)
```

```python
print(len(rating_given), len(review_summ), len(full_review))
```

```python
import pandas as pd

df = pd.DataFrame({'Rating': rating_given, 'Review Summary':review_summ, 'Full Review': full_review })

df
```

```python
driver = webdriver.Chrome()

driver.get('https://www.azquotes.com/') #Opening the amazon page page on automated chrome browser
```

```python
search = driver.find_element(By.XPATH, '//div[@class="mainmenu"]/ul/li[5]/a')
search.click()
```

```python
quote = []
author = []
quotestype = []


start = 0
end = 10
for page in range(start, end):
    quote_tag = driver.find_elements(By.XPATH, '//ul[@class="list-quotes"]/li/div/p/a[2]')
    for i in quote_tag:
        quote.append(i.text)
    next_button = driver.find_element(By.XPATH, '//div[@class="leftcol-inner"]/div[3]/li[12]/a') #botton to move to next page
    next_button.click()
    time.sleep(3)
    
for page in range(start, end):
    author_tag = driver.find_elements(By.XPATH, '//ul[@class="list-quotes"]/li/div/p/a[1]')
    for i in author_tag:
        author.append(i.text)
    next_button = driver.find_element(By.XPATH, '//div[@class="leftcol-inner"]/div[3]/li[12]/a') #botton to move to next page
    next_button.click()
    time.sleep(3)
    
for page in range(start, end):
    quotestype_tag = driver.find_elements(By.CLASS_NAME, 'tags')
    for i in quotestype_tag:
        quotestype.append(i.text)
    next_button = driver.find_element(By.XPATH, '//div[@class="leftcol-inner"]/div[3]/li[12]/a') #botton to move to next page
    next_button.click()
    time.sleep(3)
```

```python
print(len(quote), len(author), len(quotestype))
```

```python
df = pd.DataFrame({'Quotes': quote, 'Authors Name':author, 'Type of quotes': quotestype })

df
```

```python
driver = webdriver.Chrome()

driver.get('https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1') #Opening the amazon page page on automated chrome browser
```

```python
PMI_Name = []
PMI_Born_Dead = []
PMI_Term = []
PMI_Remark = []

#scraping Name of former Prime Ministers of India from the given page
Name_tag = driver.find_elements(By.XPATH, '//div[@class="TableData"]/table/tbody/tr/td[2]/div')
for i in Name_tag:
    PMI_Name.append(i.text)
    
#scraping Born-Dead of former Prime Ministers of India from the given page
Born_Dead_tag = driver.find_elements(By.XPATH, ' //div[@class="TableData"]/table/tbody/tr/td[3]/div')
for i in Born_Dead_tag:
    PMI_Born_Dead.append(i.text)
    
#scraping Term of office of former Prime Ministers of India from the given page    
PMI_Term_tag = driver.find_elements(By.XPATH, '//div[@class="TableData"]/table/tbody/tr/td[4]/div[1]')
for i in PMI_Term_tag:
    PMI_Term.append(i.text)
    
#scraping Remark of former Prime Ministers of India from the given page    
PMI_Remark_tag = driver.find_elements(By.XPATH, '//div[@class="TableData"]/table/tbody/tr/td[5]/div')
for i in PMI_Remark_tag:
    PMI_Remark.append(i.text)
```

```python
print(len(PMI_Name), len(PMI_Born_Dead), len(PMI_Term), len(PMI_Remark))
```

```python
df = pd.DataFrame({'Name': PMI_Name, 'Born-Dead':PMI_Born_Dead, 'Term of office': PMI_Term,'Remark': PMI_Remark })

df
```
